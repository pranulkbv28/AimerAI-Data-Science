# Pandas Basic

## Understanding Series and DataFrames

Pandas is a Python library that provides powerful tools for data analysis and manipulation. It has two main data structures:

- Series
- DataFrames

Think of them as containers to store data in an organized way. For beginners, it's helpful to think of a Series as a column of data, and a DataFrame as a whole table with rows and columns.

### What is Pandas Series?

A Series is a one-dimensional array, like a single column of data in a spreadsheet. Each element in a Series has an index (like a label) that makes it easy to find specific values.

Key Points about Series: **One-dimensional**: A Series only has one column of data, so it's simpler than a DataFrame.

**Indexed**: Each value has an index label, so you can access values by their label.

**Holds one data type**: All values in a Series are usually of the same data type (e.g., all integers, all floats, or all strings).

**When to Use Series:**

- When you need to store a single column or list of data.
- For simpler datasets where you only need one type of information, like a list of ages or prices.


```python
import pandas as pd

print(pd.__version__)
```

#### Creating a Series in Pandas:

To create a Series, we can use the pd.Series() function. Let's look at different ways to create a Series.

*Example 1: Creating a Series from a List*

```python
# Creating a series from a list of numbers

data = [10,20,30,40]
series = pd.Series(data)

print("Series with default index:\n", series)
```

Explanation:

> We used a list `[10, 20, 30, 40]` to create the Series.
> 
> Pandas automatically assigns default indices (0, 1, 2, 3).

*Example 2: Creating with a Custom index*

You can also assign custom labels (or indices) to each element in a Series

```python
# Creating a Series with custom index Labels
labels = ['a', 'b', 'c', 'd']
series = pd.Series(data, index=labels)
print("Series with custom indices:\n", series)
```

Explanation:

> The list `['a', 'b', 'c', 'd']` specifies custom labels for the Series.

#### Accessing values in a Series

You can retrieve specific values by using their index labels.

```python
print("Value at index 'b': ", series['b'])
```

### What is Pandas DataFrame?

A DataFrame is a two-dimensional data structure, like a table, where data is organized in rows and columns. Each column can have a different data type (e.g., integers, floats, strings).

*Key Points about DataFrames*:

- Two-dimensional: A DataFrame has rows and columns, so it's like a table.
- Labeled rows and columns: Rows and columns both have labels, making it easy to access data.
- Multiple data types: Each column can store different types of data (e.g., numbers, text).

*When to Use DataFrames*:

- When you need to store data in a table format, with multiple columns representing different attributes.
- For more complex datasets where each column represents a different variable or feature.

#### Creating a DataFrame in Pandas

To create a DataFrame, you can use the pd.DataFrame () function with different types of data sources.

```python
# Creating a DataFrame from a Dictionary

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 27, 22],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
print("DataFrame example:\n")
df
```

*Explanation*:

> We used a dictionary where each key (e.g., Name, Age, City) represents a column name, and each value is a list of column values.

#### Syntax for Creating DataFrames from Different Sources:

- `CSV File`: `df = pd.read csv('file.csv')`: eads data from a CSV file.
- `Excel File`: `df= pd. read excel('file.xIsx')`: Reads data from an Excel file.
- `SQL Database`: `df = pd. read_sql(query, connection)`: Reads data from a database query.

3. ***Accessing Rows Using `iloc`, `loc`, `at` and `iat`***

```python
# Creating a DataFrame from a Dictionary

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 50],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data, index=['a','b','c', 'd'])
df
```

a. `loc`: ***Label Based Indexing*** 

- loc is used for label-based indexing.
- This means you can use the labels (index or column names) to access rows and columns.
- loc is inclusive of the end index, so it will include both the start and end of the specified range.
- Syntax:

`DataFrame.loc[row_label, column_label]`

- *Example: Using loc to Select Rows and Columns by Labels*

```python
# Select a single row using label

row_a = df.loc['a']
print("Row 'a':\n", row_a)
```

```python
# Select multiple rows and a specific column by labels

rows_bc_age_city = df.loc[['b', 'c'], ['Age', 'City']]
print("\nRows 'b' and 'c' with 'Age' and 'City' columns:\n", rows_bc_age_city)
```

```python
# Select Rows based on condition

age_above_30 = df.loc[df['Age']>30, 'Name']
print("\nNames of people older than 30:\n", age_above_30)
```

Explanation:

- `loc` : Access rows by index label.
- `df.loc['a']` : Selects all columns in row 'a' (Alice's details).
- `df.loc[['b', 'c'], ['Age', 'City' ]]` : Selects rows with labels 'b' and 'c', displaying only the 'Age' and 'City' columns.
- `df. loc[df[ 'Age'] › 30, 'Name ']` : Uses a condition to select rows where 'Age' is greater than 30, displaying only the 'Name' column.

**`loc` using slice methods**

```python
df.loc['a':'b', 'Name':'Age']
```

Here, we can see that index slicing can be done to get the vales.

But, one thing to notice, in normal slicing it would've been -> `[start_index:end_index+1]`,

but with `loc`, it is -> `[start_index:end_index]`

b. `iloc`: **Position Based Indexing**

- iloc is used for integer-location-based indexing.
- This means you specify row and column positions (0-based index) rather than labels
- iloc is also inclusive of the starting index but exclusive of the ending index (like Python slicing).
- Syntax:

DataFrame. iloc[row position, column_position]

- Example: Using iloc to Access Data by Position

```python
# Creating a DataFrame from a Dictionary

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 50],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data, index=['a','b','c', 'd'])
df
```

```python
# Select a single row by position
row_0 = df.iloc[0]
print ("Row 0: \n", row_0)

# Select multiple rows and specific columns by position
rows_1_2_columns_1_2 = df.iloc[1:3, 1:3]
print("\nRows 1 to 2, Columns 1 to 2: \n", rows_1_2_columns_1_2)

# Select a specific value by row and column position
value_row_2_col_1 = df. iloc[2, 1]
print("\nValue at Row 2, Column 1:", value_row_2_col_1)
```

*Explanation*:

- `df. iloc[0]` : Selects all columns in the first row (row 0).
- `df. iloc[1:3, 1:3]` : Selects rows 1 and 2 and columns 1 and 2, inclusive of the start but exclusive of the end.
- `df. iloc[2, 1]` : Retrieves the value in the third row (row index 2) and second column (column index 1).

Here, if you have noticed, we are using the regular `slice` expressions, which is, `[start_index:end_index+1]`

#### `Filtering` Techniques in Pandas

Filtering is crucial when working with data as it allows you to select rows that match specific conditions. Here's a comprehensive look at filtering techniques in Pandas, with examples to illustrate each approach.

```python
# Sample Data

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 30, 35, 22, 28],
    'Salary': [50000, 54000, 61000, 45000, 52000]
}

df = pd.DataFrame(data)
df
```

1. ***Boolean Indexing with Conditions***:

Boolean indexing lets you filter rows by applying conditions directly to columns. This technique is useful for straightforward conditional filters.

```python
# Filter rows where Age > 25

df_age_above_25 = df[df['Age']>25]
print("\nRows with Age greater than 25:\n", df_age_above_25)
```

```python
# Filter rows with multiple conditions: Age > 25 & Salary > 50000

df_multi_condition = df[(df['Age']>25) & (df['Salary']>50000)]
print("\nRows with Age greater than 25 and Salary greater than 60000:\n", df_multi_condition)
```

Explanation:
- `df[df['Age'] > 25]` : Filters rows where the Age column is greater than 25.
- `df[ (df['Age"] > 25) & (df['Salary ] > 50000)]` :
  - Filters rows where Age is greater than 25 and Salary is more than 50,000.
  - The & operator is used to apply multiple conditions.

Here `and` functionality can be done using `&` and even `or` functionality can be done using `|`

2. ***Filtering Using isin0 for Multiple Values***:

The isin() method is ideal for filtering rows based on a set of specific values in a column.

```python
# Filter Rows where Name is either 'Alice' or 'David'

df_names = df[df['Name'].isin(['Alice', 'David'])]
print("\nRows with name 'Alice' or 'David':\n", df_names)
```

```python
# Filter rows where Age is either 24, 30 or 35

df_specific_ages = df[df['Age'].isin([24,30,35])]
print("\nRows with specific ages (24, 30 or 35):\n", df_specific_ages)
```

Explanation:
- `df[df[ Name']. isin(['Alice', 'David'|)]` :
  - Filters rows where the Name column contains either "Alice" or "David".
- `df[df[ Age'].isin([24, 30, 35])]` :
  - Filters rows where Age matches any value in `[24, 30, 35]`
  - This is useful for filtering specific sets of categorical or numeric values.

3. ***Filtering with isnull() or notnull() for Missing Data:***

Filtering based on missing data allows you to focus on rows with or without NaN values.

```python
# Sample data with NaN values

data_with_nan = {
    'Name': ['Alice', 'Bob', 'Charlie',  None, 'Eva'],
    'Age': [24, None, 35, 22, None],
    'Salary': [50000, None, 61000, 45000, 52000]
}
df_nan = pd.DataFrame(data_with_nan)
df_nan
```

```python
# Filter rows where Age is null

df_age_null = df_nan[df_nan['Age'].isnull()]
print("\nRows where Age is null:\n", df_age_null)
```

```python
# Filter rows where Salary is not null

df_salary_not_null = df_nan[df_nan['Salary'].notnull()]
print("\nRows where Salary is not null:\n", df_salary_not_null)
```

*Explanation*:

- `df_nan[df_nan['Age'].isnull()]`:
  - Filters rows where **Age** column contains NaN values
  - `isnull()` is ideal for identifying missing data
- `df_nan[df_nan['Salary'].notnull()]`:
  - Filters rows where **Salary** column has non-null values
  - `notnull()` is helpful for selecting rows with comlpete data in a column

#### `Handling Missing Data`

- Missing data is common in real-world datasets. Pandas provides functions to identify and handle missing values

1. `Identifying Missing Data`:

   - Use `isna()` or `isnull`

```python
import numpy as np

# Creating a DataFrame from a Dictionary

data = {
    'Name': ['Alice', 'Bob', np.nan],
    'Age': [24, 27, 22],
    'City': ['New York', np.nan, 'Chicago']
}
df = pd.DataFrame(data)
df.isna()
```

*Explanation*:

> `df.isna()` returns a **DataFrame** of the same size as `df`, with `True` where data is missing and `False` where it's present

2. `Filling Missing Values (Imputation)`

   - When dealing with missing values you can replace NaN values `mean`, `median` and `mode` values of the column

```python
# Creating a DataFrame from a Dictionary

df_with_nans = pd.DataFrame({'Score': [80, 90, None, 95, None]})
df_with_nans
```

```python
# Fill Nans with mean

mean_inputed = df_with_nans['Score'].fillna(df_with_nans['Score'].mean())
print("Filled with Mean:\n", mean_inputed)
```

```python
# Fill NaNs with median

median_inputed = df_with_nans['Score'].fillna(df_with_nans['Score'].median())
print("\nFilled with Median:\n", median_inputed)
```

```python
# Filled with mode

mode_inputed = df_with_nans['Score'].fillna(df_with_nans['Score'].mode())
print("\nFilled with Mode:\n", mode_inputed)
```

*Explanation*:

> - `fillna(df_with_nans['Score'].mean())`: Replaces missing values in Score with the column's mean
> - `fillna(df_with_nans['Score'].median())`: Replaces missing values in Score with the column's median
> - `fillna(df_with_nans['Score'].mode())`: Replaces missing values in Score with the column's mode

3. `Removing Rows/Columns with Missing Data`:

   - `dropna()`: Removes rows or columns with missing values

```python
df.dropna(inplace=True)
df
```

*Explaination*:

> - `dropna()` removes rows that have at least one missing value.
> - `inplace=True` makes this change directly in df.

# Advanced Pandas Techniques

```python
import pandas as pd

print(pd.__version__)
```

```python
# Sample DataFrame

data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Values': [10,20,15,25,30,35],
    'Score': [1,2,3,4,5,6]
}
df = pd.DataFrame(data)
df
```

## Simple Aggregation with a Single Function

```python
# Aggregation: Sum of 'Values' grouped by 'Category'

result = df.groupby('Category')['Values'].sum()
print(result)
```

*Explanation*:

- The `groupby` function groups data by **Category** (A,B,C).
- The `sum` functions adds up the **Values** for each group:
  - **A**: 10+20 = 30
  - **B**: 15+25 = 40
  - **A**: 30+36 = 65

## Multiple Aggregations on a Single Column

```python
# Aggregation: Applying 'sum' and 'mean' to 'Values'

result = df.groupby('Category')['Values'].agg((['sum', 'mean']))
print(result)
```

*Explanation*:

- The `groupby('Category')` groups the data by the 'Category' column.
- The `agg([' sum', "mean' ])` applies:
  - sum to the 'Values' column, adding up the values for each group.
  - mean to the 'Values' column, calculating the average value for each group.

## Aggregating Multiple Columns

```python
# Aggregation: Applying different function to 'Values' and 'Score'

result = df.groupby('Category').agg({'Values': 'sum', 'Score': 'mean'})
print(result)
```

## Custom Aggregation Functions

```python
# Aggregation: Custom function to calculate range (max - min) of 'Values'

result = df.groupby('Category')['Values'].agg(lambda x: x.max() - x.min())
print(result)
```

*Code Explanation*:
- `groupby ('Category')`: Groups the data by the 'Category' column.
- `agg(lambda x: x.max() - x.min())`: Applies a custom aggregation function to the 'Values' column that calculates the range (max value - min value) for each group.

*Output Explanation*:
- **A**: The range of values for category A is `20 (maximum) - 10 (minimum) = 10`.
- **B**: The range of values for category B is `25 (maximum) - 15 (minimum) = 10`.
- **C**: The range of values for category C is `35 (maximum) - 30 (minimum) = 5`.
